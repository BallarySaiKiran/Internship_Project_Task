```python
# Event Cost Estimator & Budget Negotiation Bot

print("\nWelcome to Event Cost Estimator!")
# Taking budget from user
event_budget = int(input("\nEnter your total event budget (₹): "))

# Available services
services = {
    "hall": 40000,
    "decoration": 20000,
    "catering": 50000,
    "photographer": 30000,
    "lighting": 15000,
    "entertainment": 10000
}

total_cost = 0
selected_services = []


def display_services():
    print("\nSelected Services:")
    for service in selected_services:
        print(f"- {service}: ₹{services[service]}")
    print(f"Total Estimated Cost: ₹{total_cost}")

while True:
    service = input("Enter a service to include (e.g., catering, photographer,....:").strip().lower()

    if service in services:
        total_cost += services[service]
        selected_services.append(service)
        print(f"'{service}' added to your event plan.")
        display_services()
    else:
        print(f"Service '{service}' is not available.")

    more=input("\nDo you want to add another service? (yes/no): ").strip().lower()
    if more!="yes":
        break

display_services()

# Budget negotiation logic
while total_cost>event_budget:
    print("The total cost exceeds your budget.")
    remove=input("\nWould you like to remove a service to reduce the cost? (yes/no): ").strip().lower()
    if remove=="yes":
        to_remove=input("Enter the service you want to remove: ").strip().lower()
        if to_remove in selected_services:
            total_cost-=services[to_remove]
            selected_services.remove(to_remove)
            print(f"'{to_remove}' removed from the plan.")
            display_services()
        else:
            print("That service is not in your current selection.")
    else:
        print("You chose not to remove any services.")
        break
        
print("\nThank you for using our event cost estimator. We wish you a successful event!")
```
